**Q1. What is the meaning of multiple inheritance?**

Multiple inheritance is a feature in object-oriented programming
languages that allows a class to inherit attributes and methods from
more than one parent class (superclass). In multiple inheritance, a
single derived class (subclass) can inherit and combine the features of
two or more base classes. This means that the subclass has access to the
attributes and methods of all its parent classes.

**Q2. What is the concept of delegation?**

Delegation in Python is a design pattern that allows one object to
forward or delegate specific tasks or responsibilities to another
object. It is a fundamental concept in object-oriented programming (OOP)
and is closely related to the principles of composition and code reuse.
Delegation enables objects to collaborate, divide work efficiently, and
maintain a high degree of modularity in the codebase.

**Q3. What is the concept of composition?**

Composition is a fundamental concept in object-oriented programming
(OOP) that refers to the practice of building complex objects by
combining simpler or more basic objects. It involves creating a new
object (the composed object) by including other objects (the component
objects) as parts or attributes. Composition allows you to create
complex structures and behaviors while promoting code reuse,
flexibility, and maintainability.

Bound methods are a concept in Python that refers to methods attached to
instances of a class. When you access a method from an instance of a
class, it becomes a bound method, and it is automatically passed the
instance as its first argument (usually named self). This allows the
method to operate on the data or attributes specific to that instance.

Here's how bound methods work and how to use them:

1.  **Bound Method Creation**:

-   When you access a method from an instance, Python automatically
    > binds the method to that instance. The instance is passed as the
    > first argument (self by convention) to the method.

1.  **Accessing Instance Attributes**:

-   Bound methods have access to the instance's attributes and data.
    > They can read and modify the instance's data as needed.

1.  **Self-Reference**:

-   Within a bound method, you can refer to the instance using the self
    > parameter, which allows you to work with instance-specific data
    > and behaviour.

**Q5. What is the purpose of pseudoprivate attributes?**

Pseudoprivate attributes in Python, denoted by a double underscore
prefix, serve several important purposes in object-oriented programming.
Firstly, they implement name mangling, automatically modifying the
attribute name to include the class name, which helps prevent naming
conflicts when multiple classes are involved in inheritance or
composition. Secondly, they act as a signal to other developers,
indicating that these attributes are intended for internal use within
the class and should not be directly accessed from outside. While they
don't provide strict access control, they encourage encapsulation and
discourage unintended modification. Moreover, pseudoprivate attributes
enhance code clarity and organization by following a naming convention,
making it easier to distinguish between attributes meant for public use
and those for internal implementation. Inheritance benefits from
pseudoprivate attributes as they avoid attribute name collisions in
subclasses. Additionally, pseudoprivate attributes help in avoiding name
clashes in large codebases and serve as a form of documentation,
conveying the intended usage of the attributes. Overall, pseudoprivate
attributes contribute to code maintainability, reducing the risk of
unintended changes to internal attributes and helping to maintain the
expected behaviour of a class.
