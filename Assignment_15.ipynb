1.  **What are the new features added in Python 3.8 version?**

**Assignment Expressions (The Walrus Operator)**:

-   The introduction of the **:=** operator allows assignment
    expressions within expressions. It helps simplify code by allowing
    variable assignment within expressions.

**Positional-Only Parameters**:

-   Python 3.8 introduced the ability to specify positional-only
    parameters in function definitions using **/** syntax. This allows
    functions to enforce how arguments are passed.

**f-strings Improvements**:

-   Python 3.8 introduced the **=** specifier in f-strings to display
    variable names and their values, aiding in debugging.

**Syntax Warnings**:

-   Python 3.8 introduced SyntaxWarnings for suspicious syntax that may
    lead to runtime errors in later Python versions.

**Math Functions**:

-   The **math** module added new functions like **math.prod()** for
    calculating the product of iterable elements.

1.  **What is monkey patching in Python?**

Monkey patching in Python refers to the practice of dynamically
modifying or extending the behavior of existing classes, modules, or
functions at runtime. It involves making changes to code that is already
loaded and running, typically without modifying the original source
code. Monkey patching can be a powerful technique, but it should be used
with caution as it can lead to maintenance challenges and unexpected
behavior if not done carefully.

1.  **What is the difference between a shallow copy and deep copy?**

In Python, a "shallow copy" and a "deep copy" are two different ways of
duplicating objects, primarily lists and dictionaries. These terms
describe how nested objects (objects within objects) are copied when you
clone a container object. Here are the key differences between them:

**Shallow Copy**:

1.  **Copying the Outer Object**:

> In a shallow copy, the outermost container (list, dictionary, etc.) is
> duplicated, creating a new object. However, the inner objects
> (elements or values) are still references to the same objects as the
> original.

1.  **Nested Objects are Shared**:

> Any changes made to the inner objects in the copied container will
> also affect the original container. This is because they are
> references to the same objects in memory.

1.  **Shallow Copy Methods**:

> In Python, you can create shallow copies using the **copy.copy()**
> function (or the **.copy()** method for lists) for lists and
> dictionaries.

**Deep Copy**:

1.  **Copying the Outer and Inner Objects**:

> In a deep copy, both the outer object and all the inner objects are
> duplicated recursively. This means that a new container and new copies
> of all nested objects are created.

1.  **Nested Objects are Independent**:

> Any changes made to the inner objects in the copied container will not
> affect the original container. They are completely independent
> objects.

1.  **Deep Copy Methods**:

> In Python, you can create deep copies using the **copy.deepcopy()**
> function (from the **copy** module) for lists and dictionaries.

1.  **What is generator comprehension?**

> Generator comprehension, also known as generator expression, is a
> concise way to create a generator in Python. It is similar to a list
> comprehension but generates values lazily on-the-fly, one at a time,
> instead of creating a complete list in memory. This can be especially
> useful when dealing with large datasets or when you want to avoid
> unnecessary memory usage.
>
> The syntax for a generator comprehension is similar to a list
> comprehension, but it uses parentheses **()** instead of square
> brackets **\[\]**. Here's the basic structure:

generator_expression = (expression for element in iterable if condition)

-   **expression** is the expression used to generate values.

-   **element** is the variable that takes on each item from the
    iterable.

-   **iterable** is the source of data you want to iterate over.

-   **condition** (optional) is an expression that filters which items
    are included in the generator.

Here's an example that creates a generator to generate squares of
numbers from 0 to 9:

squares_generator = (x\*\*2 for x in range(10))

\# To iterate and print the squares

for square in squares_generator:

print(square)

The key advantage of using generator comprehensions is that they are
memory-efficient. Values are generated one at a time as you iterate over
the generator, so it doesn't store the entire sequence in memory, making
it suitable for large datasets or infinite sequences.
