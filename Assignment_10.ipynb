**Q1. What is the difference between \_\_getattr\_\_ and
\_\_getattribute\_\_?**

In Python, the main difference between \_\_getattr\_\_ and
\_\_getattribute\_\_ is that \_\_getattr\_\_ is used when an attribute
is not found. \_\_getattribute\_\_ is used before looking at the actual
attributes on the object.

\_\_getattribute\_\_ is called unconditionally when an attribute is
being retrieved from an instance.

\_\_getattr\_\_ is called only when the attribute was not found.

If a class also defines \_\_getattr\_\_(), the latter will not be called
unless \_\_getattribute\_\_() either calls it explicitly or raises an
AttributeError.

If \_\_getattribute\_\_ needs to raise an AttributeError exception,
nothing happens and \_\_getattr\_\_ takes over.

**Q2. What is the difference between properties and descriptors?**

The main difference between properties and descriptors is that
properties are a high-level application of descriptors. Descriptors are
a low-level mechanism that lets you hook into an object's attributes
being accessed. Properties are implemented using descriptors.

Properties are objects that have a getter, setter, and deleter method.
The getter method is called when the property is accessed, the setter
method is called when the property is set, and the deleter method is
called when the property is deleted.

Descriptors are objects that implement the descriptor protocol. The
descriptor protocol has three methods: \_\_get\_\_(), \_\_set\_\_(), and
\_\_delete\_\_(). The \_\_get\_\_() method is called when the descriptor
is accessed, the \_\_set\_\_() method is called when the descriptor is
set, and the \_\_delete\_\_() method is called when the descriptor is
deleted.

Properties are a type of descriptor. When a property is assigned to an
attribute, the property object becomes the descriptor for that
attribute.
