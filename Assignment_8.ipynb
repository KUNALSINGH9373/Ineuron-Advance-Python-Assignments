**Q2. How are class-based exceptions that have been raised matched to
handlers?**

In Python, class-based exceptions that have been raised are matched to
handlers using a mechanism known as exception handling or exception
propagation. This process involves the interpreter searching for the
appropriate exception handler in the call stack.

Here's how it works:

1.  **Raising an Exception**: When an exception is raised within your
    > code, either by using the raise statement or as a result of an
    > error condition (e.g., division by zero), Python creates an
    > exception object that carries information about the exception,
    > including its type (which is typically a class representing the
    > exception) and any associated data or messages.

2.  **Propagation of the Exception**: Python starts looking for an
    > appropriate exception handler for the raised exception. It does
    > this by unwinding the call stack, which means it goes up the chain
    > of function and method calls in reverse order to find a matching
    > except block.

3.  **Matching the Exception Type**: At each level of the call stack,
    > Python checks the type of the raised exception against the types
    > specified in except blocks. It looks for an except block with a
    > matching exception type. If a match is found, the code inside that
    > except block is executed.

4.  **Handling the Exception**: Once a matching except block is found
    > and executed, the exception is considered handled, and Python does
    > not continue searching for more matching handlers. The code within
    > the matching except block can handle the exception in various
    > ways, such as logging the error, displaying an error message, or
    > taking corrective action.

5.  **Exception Propagation if Not Matched**: If Python reaches the top
    > of the call stack (the global level) without finding a matching
    > except block, the program will terminate, and an unhandled
    > exception error message will be displayed. This means that
    > unhandled exceptions typically result in program termination.
