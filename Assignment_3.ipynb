**1. What is the concept of an abstract superclass?**

An abstract superclass (or abstract base class) is a class in
object-oriented programming that is meant to serve as a blueprint or
template for other classes but is not intended to be instantiated
directly. Instead, it is designed to be subclassed by other classes,
which inherit its attributes and methods.

from abc import ABC, abstractmethod

class Shape(ABC): \# Abstract superclass

@abstractmethod

def area(self):

pass

class Circle(Shape):

def \_\_init\_\_(self, radius):

self.radius = radius

def area(self):

return 3.14 \* self.radius \* self.radius

class Square(Shape):

def \_\_init\_\_(self, side_length):

self.side_length = side_length

def area(self):

return self.side_length \* self.side_length

Shape with abstract method area

**2. What happens when a class statement's top level contains a basic
assignment statement?**

In Python, when a class statement's top level contains a basic
assignment statement, that assignment statement creates a class-level
attribute (also known as a class attribute) rather than an
instance-level attribute. This class attribute is shared among all
instances of the class and is associated with the class itself rather
than individual instances.

**3. Why does a class need to manually call a superclass's \_\_init\_\_
method?**

You may need to call the superclass's \_\_init\_\_ method manually in
the subclass's \_\_init\_\_ method for several important reasons:

1.  **Initialization of Superclass Attributes**: The \_\_init\_\_ method
    > of the superclass is responsible for initializing the attributes
    > specific to that class. If you don't call the superclass's
    > \_\_init\_\_ method, those attributes won't be initialized
    > properly, and the subclass may not function as expected.

2.  **Inheritance of Constructor Logic**: By calling the superclass's
    > \_\_init\_\_ method, you ensure that any logic or initialization
    > performed in the superclass's constructor is executed. This
    > includes any necessary setup, default values, or validation that
    > the superclass's constructor may provide.

1.  **How can you augment, instead of completely replacing, an inherited
    method?**

> To augment (extend or enhance) an inherited method in a subclass
> without completely replacing it, you can follow these steps:

1.  **Call the Parent Class Method**: Inside the subclass's method,
    > first, call the parent class's method using super(). This allows
    > you to execute the behavior defined in the parent class.

2.  **Add Your Custom Logic**: After calling the parent class's method,
    > you can add your own custom logic to extend or modify the behavior
    > of the method.

Here's an example to illustrate how to augment an inherited method:

class Animal:

def speak(self):

return "Animal speaks"

class Dog(Animal):

def speak(self):

\# Call the parent class's method using super()

parent_speak = super().speak()

\# Add custom behavior to extend the method

return f"{parent_speak} and barks"

\# Create instances

animal = Animal()

dog = Dog()

\# Call the speak method for both instances

print(animal.speak()) \# Output: Animal speaks

print(dog.speak()) \# Output: Animal speaks and barks

1.  **How is the local scope of a class different from that of a
    function?**

<!-- -->

1.  **Scope Type**:

-   **Function Scope**: The local scope of a function is created when
    > the function is called and destroyed when the function exits. It
    > is temporary and exists only during the execution of the function.

<!-- -->

-   **Class Scope**: The local scope of a class exists for the entire
    > lifetime of the class. It is created when the class is defined and
    > continues to exist as long as the program runs. It is not tied to
    > the creation of class instances.

1.  **Accessibility**:

-   **Function Scope**: Variables defined within a function's local
    > scope are accessible only within that function. They are not
    > visible or accessible outside the function.

<!-- -->

-   **Class Scope**: Variables defined within a class's local scope
    > (i.e., class attributes) are accessible both within the class and
    > from instances of the class. They are shared among all instances
    > of the class and can be accessed using the class name.

1.  **Lifetime**:

-   **Function Scope**: Local variables in a function are created when
    > the function is called and destroyed when the function exits. They
    > have a short lifetime and are typically used for temporary storage
    > during the function's execution.

<!-- -->

-   **Class Scope**: Class attributes exist for the entire lifetime of
    > the class. They are created when the class is defined and persist
    > as long as the program is running. They store data or behavior
    > shared among all instances of the class.
